```python
menu = {
    "1": {"name": "Tandoori chicken (4 pieces)", "quantity": "4 pieces", "price": 240, "discount": 0},
    "2": {"name": "Vegan Burger (1 piece)", "quantity": "1 piece", "price": 320, "discount": 0},
    "3": {"name": "Truffle cake (500gm)", "quantity": "500gm", "price": 900, "discount": 0}
}
users = {}
user_orders = {}

def generate_food_id():
    return f"F{len(menu) + 1:03d}"

def register():
    print("User Registration")
    full_name = input("Full Name: ")
    phone = input("Phone: ")
    email = input("Email: ")
    address = input("Address: ")
    password = input("Password: ")

    users[email] = {
        "full_name": full_name,
        "phone": phone,
        "address": address,
        "password": password
    }
    print("Registration successful!")

def login():
    print("User Login")
    email = input("Email: ")
    password = input("Password: ")

    if email in users and users[email]["password"] == password:
        print(f"Welcome, {users[email]['full_name']}!")
        return email
    else:
        print("Login failed. Please check your credentials.")
        return None

def place_order(user_email):
    print("Place New Order")
    selected_items = []

    while True:
        print("\nMenu:")
        for item_id, item_info in menu.items():
            print(f"{item_id}: {item_info['name']} [{item_info['quantity']}] [INR {item_info['price']}]")

        order_item = input("Enter the item number you want to order (0 to finish): ")
        if order_item == "0":
            break
        if order_item in menu:
            selected_items.append(order_item)
        else:
            print("Invalid item number. Please select a valid item from the menu.")

    order_details = {"items": []}
    total_price = 0

    for item_id in selected_items:
        item_info = menu[item_id]
        order_details["items"].append(item_info)
        total_price += item_info["price"]

    user_orders[user_email].append(order_details)
    print("Order placed successfully!")
    print("Selected Items:")
    for item_info in order_details["items"]:
        print(f"{item_info['name']} [{item_info['quantity']}] [INR {item_info['price']}]")
    print(f"Total Price: INR {total_price}")

def order_history(user_email):
    print("Order History for User:", users[user_email]["full_name"])
    if user_email in user_orders:
        for idx, order in enumerate(user_orders[user_email], 1):
            print(f"Order {idx}:")
            for item_info in order["items"]:
                print(f"{item_info['name']} [{item_info['quantity']}] [INR {item_info['price']}]")
            print()

def main():
    while True:
        print("\nFood Order App")
        print("1. Register")
        print("2. Login")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            register()
        elif choice == "2":
            user_email = login()
            if user_email:
                user_orders.setdefault(user_email, [])
                while True:
                    print("User Options:")
                    print("1. Place New Order")
                    print("2. Order History")
                    print("3. Update Profile")
                    print("4. Log Out")
                    user_choice = input("Enter your choice: ")

                    if user_choice == "1":
                        place_order(user_email)
                    elif user_choice == "2":
                        order_history(user_email)
                    elif user_choice == "3":
                        # Implement profile update
                        print("Profile update is not implemented in this example.")
                    elif user_choice == "4":
                        print("Logged out.")
                        break
                    else:
                        print("Invalid choice. Please select a valid option.")
        elif choice == "3":
            print("Thank you for using the Food Order App. Goodbye!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
